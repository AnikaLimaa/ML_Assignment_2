```python
## 7. Write a program to sum all prime numbers within a range of 10 to 1000. 

import math

s=0

for i in range(10,1000+1):
    
    
    n=math.sqrt(i)
    n=int(n)
    
    f=0
    
    for j in range(2,n+2):
        
        if i%j==0:
            f=1
            break
            
    if f==0:
        s=s+i
        
print(s)
```
